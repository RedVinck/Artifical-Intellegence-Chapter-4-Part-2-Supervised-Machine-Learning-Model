```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#Read data
data = pd.read_excel('data.xlsx', sheet_name='PS19A_PERSON__SmplR10K')

#Shape the data
data.shape
```

```python
#Brief overview of 5 rows
data.head(1)
```

```python
#Fill all empty values with 0 for feature columns
data['person_name'] = data['person_name'].fillna(0)
data['person_address'] = data['person_address'].fillna(0)
data['person_ctry_code'] = data['person_ctry_code'].fillna(0)

#Fill all empty values with 0 for predicted columns
data['country'] = data['country'].fillna(0)
data['city'] = data['city'].fillna(0)
data['region'] = data['region'].fillna(0)
data['postal'] = data['postal'].fillna(0)
data['area'] = data['area'].fillna(0)
data['street'] = data['street'].fillna(0)
data['co'] = data['co'].fillna(0)
data['building'] = data['building'].fillna(0)
data['nr'] = data['nr'].fillna(0)
```

```python
#Brief overview of first 5 rows
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
l3 = LabelEncoder()

person_name = data.person_name.values
person_name = l3.fit_transform(person_name)

person_address = data["person_address"].values.astype(str)
person_address = l3.fit_transform(person_address)

person_ctry_code = data.person_ctry_code.values
person_ctry_code = l3.fit_transform(person_ctry_code)

#Make the columns as string and encode it to a numeric
country = data.country.values.astype(str)
country = l3.fit_transform(country)

region = data.region.values.astype(str)
region = l3.fit_transform(region)

city = data.city.values.astype(str)
city = l3.fit_transform(city)

postal = data.postal.values.astype(str)
postal = l3.fit_transform(postal)

area = data.country.values.astype(str)
area = l3.fit_transform(area)

nr = data.nr.values.astype(str)
nr = l3.fit_transform(nr)

street = data.street.values.astype(str)
street = l3.fit_transform(street)

building = data.building.values.astype(str)
building = l3.fit_transform(building)

co = data.co.values.astype(str)
co = l3.fit_transform(co)
```

```python
data['person_name'] = person_name
data['person_address'] = person_address
data['person_ctry_code'] = person_ctry_code

#Replace originalcolumns witht the numeric columns
data['country'] = country
data['region'] = region
data['city'] = city
data['postal'] = postal
data['area'] = area
data['nr'] = nr
data['street'] = street
data['building'] = building
data['co'] = co
```

```python
#Brief overview of first 5
data.head(5)
```

```python
#Detailed overview
data.describe()
```

```python
from sklearn.model_selection import train_test_split

#List columns to predict and input columns
predicted_class = ['co', 'building', 'street', 'nr', 'area', 'postal', 'city', 'region', 'country']
feature_columns = ['person_name', 'person_address', 'person_ctry_code']
```

```python
#Gather input data and data to predict
X = data[feature_columns].values
Y = data[predicted_class].values

#Create training and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3) #70% training and 30% test
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```
