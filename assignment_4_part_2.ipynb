# Preprocess

```python
import pandas as pd

# Read data
data = pd.read_excel('data.xlsx', sheet_name = 'PS19A_PERSON__SmplR10K')

# Shape data
data.shape
```

```python
# Brief overview of 5 rows
data.head(5)
```

```python
import re

# Count commas
def countCommas(a):
    return str(a.person_address).count(',')

# Count numbers
def countNumbers(a):
    return len(re.findall(r"[0-9]+", str(a.person_address)))

# Count words
def countWords(a):
    return len(re.findall(r"[a-zA-Z]+", str(a.person_address)))

# Count characters
def countCharacters(a):
    return len(re.findall(r"[^ ]", str(a.person_address)))

data['comma_count'] = data.apply(countCommas, axis = 1)
data['number_count'] = data.apply(countNumbers, axis = 1)
data['word_count'] = data.apply(countWords, axis = 1)
data['character_count'] = data.apply(countCharacters, axis = 1)
```

```python
# Brief overview of 5 rows
data.head(5)
```

# Train

```python
from sklearn.model_selection import train_test_split

# List input and output columns
input_columns = ['comma_count', 'number_count', 'word_count', 'character_count']
output_columns = ['label']

# Gather input and output data
X = data[input_columns].values
Y = data[output_columns].values

# Create training and test set, 2/3 training and 1/3 test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 12)

# Train the model using the training set
knn.fit(X_train, Y_train)
```

# Deploy

```python
from sklearn import metrics

# Predict labels for the test set
Y_pred = knn.predict(X_test)

# Print model accuracy, precision and recall
print('Accuracy:', metrics.accuracy_score(Y_test, Y_pred))
print('Precision:', metrics.precision_score(Y_test, Y_pred))
print('Recall:', metrics.recall_score(Y_test, Y_pred))
```
