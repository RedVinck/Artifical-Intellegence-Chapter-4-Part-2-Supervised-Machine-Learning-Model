```python
#Read data
data = pd.read_excel('data.xlsx')
```

```python
#Fill all empty values with 0
data['country'] = data['country'].fillna(0)
data['city'] = data['city'].fillna(0)
data['region'] = data['region'].fillna(0)
data['postal'] = data['postal'].fillna(0)
data['area'] = data['area'].fillna(0)
data['street'] = data['street'].fillna(0)
data['co'] = data['co'].fillna(0)
data['building'] = data['building'].fillna(0)
data['nr'] = data['nr'].fillna(0)
```

```python
#Brief overview of first 5 rows
data.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
l3 = LabelEncoder()

#make the columns as string and encode it to a numeric
country = l3.fit_transform(data['country'].astype(str))
city = l3.fit_transform(data['city'].astype(str))
region = l3.fit_transform(data['region'].astype(str))
postal = l3.fit_transform(data['postal'].astype(str))
area = l3.fit_transform(data['area'].astype(str))
street = l3.fit_transform(data['street'].astype(str))
co = l3.fit_transform(data['co'].astype(str))
building = l3.fit_transform(data['building'].astype(str))
nr = l3.fit_transform(data['nr'].astype(str))
```

```python
#Replace originalcolumns witht the numeric columns
data['city'] = city
data['country'] = country
data['region'] = region
data['postal'] = postal
data['area'] = area
data['street'] = street
data['co'] = co
data['building'] = building
data['nr'] = nr
```

```python
#Brief overview of first 50
data.head(50)
```

```python
#Detailed overview
data.describe()
```

```python
#Just checkking how many are "right" and "false"
correct_address_map = {1.0: 1, 0.0: 0}
data['label'] = data['label'].map(correct_address_map)
correct_address_true_count = len(data.loc[data['label'] == True])
correct_address_false_count = len(data.loc[data['label'] == False])
(correct_address_true_count,correct_address_false_count)
```

```python
#renaming columns
from sklearn.model_selection import train_test_split
feature_columns = ['co', 'building', 'street','nr','area','postal','city','region','country']
predicted_class = ['label']
```

```python
#renaming columns
X = data[feature_columns].values
y = data[predicted_class].values


#renaming columns
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=10)
```

```python
from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values=0, strategy="mean")

X_train = fill_values.fit_transform(X_train)
X_test = fill_values.fit_transform(X_test)
```

```python
#HELP HAHAH
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

# Load iris dataset from sklearn
iris = datasets.load_iris()

# Declare an of the KNN classifier class with the value with neighbors.
knn = KNeighborsClassifier(n_neighbors=6)

# Fit the model with training data and target values
knn.fit(X_train, iris['target'])

# Provide data whose class labels are to be predicted
X = [
    [5.9, 1.0, 5.1, 1.8],
    [3.4, 2.0, 1.1, 4.8],
]

# Prints the data provided
print(X_test )

# Store predicted class labels of X
prediction = knn.predict(X_test )

# Prints the predicted class labels of X
print(prediction)
```
